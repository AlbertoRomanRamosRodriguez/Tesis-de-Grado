```python
from os import path, listdir, removedirs, rename
from utilities.dataset import zipDataset, getTrainingAndTest
from utilities.general import checkPathExists
```

```python
# Showing a list of datasets on the local directory
databases = [f for f in listdir() if f.isupper()]
for i, d in enumerate(databases):
    print(f"{i} - {d}")
    
#BD filesystem

BD_PATH = path.join(databases[int(input("\nSelect Database: "))])
print(f'Selected Database {BD_PATH}')

del databases

IMAGES = listdir(BD_PATH)

for f in IMAGES:
    if len(f.split(' ')) > 1:
        new_f = ''.join(f.split(' '))
        rename(path.join(BD_PATH, f), path.join(BD_PATH, new_f))

# Dataset filesystem

DATASET_PATH = path.join('Dataset')
checkPathExists(DATASET_PATH)

TRAIN_PATH = path.join(DATASET_PATH, 'train')
checkPathExists(TRAIN_PATH)
TRAIN_IMG_PATH = path.join(TRAIN_PATH, 'images')
checkPathExists(TRAIN_IMG_PATH)
TRAIN_MASK_PATH = path.join(TRAIN_PATH, 'masks')
checkPathExists(TRAIN_MASK_PATH)

TEST_PATH = path.join(DATASET_PATH, 'test')
checkPathExists(TEST_PATH)
TEST_IMG_PATH = path.join(TEST_PATH, 'images')
checkPathExists(TEST_IMG_PATH)
TEST_MASK_PATH = path.join(TEST_PATH, 'masks')
checkPathExists(TEST_MASK_PATH)
```

```python
IMAGES
```

```python
import re

masks = [img for img in IMAGES if img.endswith('_BDP.tif')]
im_names = {''.join(m.split('_BDP')):m for m in masks}

im_names
```

```python
import shutil

train_set, test_set = getTrainingAndTest(sorted(im_names), len(im_names), TRAIN_IMG_PATH, 0.7)
```

```python
test_set
```

```python
for im in train_set:
    image_s = path.join('.',BD_PATH,im)
    mask_s = path.join('.',BD_PATH,im_names[im])
    
    image_d = path.join(TRAIN_IMG_PATH,im)
    mask_d = path.join(TRAIN_MASK_PATH,im_names[im])
    
    shutil.copyfile(image_s, image_d)
    shutil.copyfile(mask_s, mask_d)

for im in test_set:
    image_s = path.join('.',BD_PATH,im)
    mask_s = path.join('.',BD_PATH,im_names[im])
    
    image_d = path.join(TEST_IMG_PATH,im)
    mask_d = path.join(TEST_MASK_PATH,im_names[im])
    
    shutil.copyfile(image_s, image_d)
    shutil.copyfile(mask_s, mask_d)
```
